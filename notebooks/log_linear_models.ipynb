```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Linear Model

```python
x = np.arange(-3,3,0.1)
b = 2
c = 1
y = b*x + c
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Log Linear Model

```python
x = np.arange(-3,3,0.1)
b = 2
c = 1
log_y = b*x + c
y = np.exp(log_y)

plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Above has to be log linear model. We are taking exponential of y and hence it is tipping up. Below is also log linear. Compare this with linear log below where b < 0. Log linear is more popular than linear log in practice.

```python
x = np.arange(-3,3,0.1)
b = -2
c = 1
log_y = b*x + c
y = np.exp(log_y)

plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Linear Log Model

```python
x = np.arange(-3,3,0.1)
b = 2
c = 1
y = b*np.log(x) + c


plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Above has to be linear log because y is getting flatten out. Below is also linear log as y is flatten out.

```python
x = np.arange(-3,3,0.1)
b = -2
c = 1
y = b*np.log(x) + c


plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Log Log Model

```python
x = np.arange(-3,3,0.1)
b = 2
c = 1
log_y = b*np.log(x) + c
y = np.exp(log_y)


plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
